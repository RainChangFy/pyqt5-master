```python
import pandas as pd
```

```python
txt_path ="D:/danmu.txt"

data_table = pd.read_csv(txt_path,sep=',',names= ['start_point','b','c','time','d','e','f','g','comments'])
```

```python
# 把 文本写出 到 txt 文件中
f = open('D:/ciyun.txt','w+',encoding = 'utf-8')

text  = data_table.comments.to_list()
text  = str(text).strip('[').strip(']')

f.write(text)
f.close()
```

```python
import datetime
```

```python
# 对我们有用的信息只有三列
new_tables = data_table[['start_point','d','comments']]
new_tables
```

```python
# 将时间戳转化为日期
new_tables['comment_times'] = new_tables.d.apply(lambda x :datetime.datetime.fromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
new_tables
```

```python
# 将 start_point 秒 转化 为 分秒
new_tables['times_point'] = new_tables.start_point.apply(lambda x: datetime.timedelta(seconds =int(x)))
```

```python
new_tables
```

```python
tranform_table = new_tables
tranform_table['times_point'] = tranform_table.times_point.apply(lambda x:str(x)[3:])
tranform_table['date_days'] = tranform_table.comment_times.apply(lambda x:datetime.datetime.fromisoformat(x).strftime("%Y-%m-%d"))
tranform_table['date_hours_'] = tranform_table.comment_times.apply(lambda x:datetime.datetime.fromisoformat(x).strftime("%H:%M:%S"))
tranform_table
```

```python
# 首先，以 d 为准进行排序
tranform_table.sort_values(by ='d')
```

```python
# 以评论时间顺序进行绘图

times_count = tranform_table.sort_values(by ='d')

times_count['Date'] = pd.to_datetime(times_count.comment_times)
times_count = times_count.set_index("Date")
times_count = times_count.resample('30T').count()
times_count


# times_count = tranform_table.group('comment_times').resample('30S').sum()
```

```python
times_count['count'] = times_count['start_point'].cumsum()
```

```python
times_count
```

```python
from pyecharts import options as opts
from pyecharts.charts import Line
```

```python
# str(times_count.index.to_list()[0].isoformat())
a = times_count['count'].to_list()
```

```python
# 绘制曲线走势图

c = (
    Line()
    .add_xaxis([str(i.isoformat()) for i in times_count.index.to_list()])
    .add_yaxis('评论数量',a,is_smooth=True,color = 'blue')
    .set_series_opts(
        areastyle_opts=opts.AreaStyleOpts(opacity=0.5),
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="《【何同学】我拍了一张600万人的合影...》时间线评论数量变化"),
        xaxis_opts=opts.AxisOpts(
            axistick_opts=opts.AxisTickOpts(is_align_with_label=True),
            is_scale=False,
            boundary_gap=False,
            
            name_rotate = 45
        
        ),
        
        legend_opts = opts.LegendOpts(pos_left = 'right')
        
        
    )
    .render("D:/line_areastyle_boundary_gap.html")
)
```

```python
# 绘制时间线
from pyecharts.charts import Bar, Timeline
from pyecharts.faker import Faker
import random

tl = Timeline()
for i in date_list:
    bar = (
        Bar()
        .add_xaxis([str(i.isoformat()).split('T')[-1] for i in trans[trans['date_days']==str(i)].resample('30T').count().index.to_list()])
        .add_yaxis("Comment Count", trans[trans['date_days']==str(i)].resample('30T').count().d.to_list(),color = random.choice(['orange','blue','red']))
        
        .set_global_opts(title_opts=opts.TitleOpts("{}评论时间分布图".format(i)))
    )
    tl.add(bar, "{}年".format(str(i)))
tl.render("D:/timeline_bar.html")
```
